```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json("../data/tables/contrataciones_arr.json")
```

```python
def number_tenderers(x):
    if 'numberOfTenderers' in x:
        return x['numberOfTenderers']
    elif 'tenderers' in x:
        return len(x['tenderers'])
    else:
        return np.nan
```

```python
df['NT'] = df['tender'].apply(number_tenderers)
```
